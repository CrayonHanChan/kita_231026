```python
# fare_cat 파생변수
def fare_cat(f):
    if f <= 15:
        cat = 4
    elif f <= 30:
        cat = 3
    elif f <= 100:
        cat = 2
    else:
        cat = 1
    return cat

titanic_df['fare_cat'] = titanic_df['fare'].apply(fare_cat)
titanic_df.head()
```

- 타이타닉에서 구조순서가 나이가 어린사람, 그다음 여성, 부자인가아닌가, 마지막에는 결국 가난하고 남자인사람이 생존률이 가장 적을것
- 즉, 여자라면 구조순위에서 남자보다 앞서니까 생존률이 높고 나이가 어릴수록 구조순위에서 나이가 많은사람보다 앞서니까 생존률이 높을것으로 추정하고 만들어보기. 

```python
# 성별로 구분한 나이별 파생변수
def get_combined_cat(sex, age):
    cat = ''
    if age <= 10: 
        cat = 'child'
    elif sex == 1:  # 여성인 경우
        if age <= 20: 
            cat = 'female_teen'
        elif age <= 60:
            cat = 'female_adult'
        else: 
            cat = 'female_elder'
    else:  # 남성인 경우
        if age <= 20:
            cat = 'male_teen'
        elif age <= 60:
            cat = 'male_adult'
        else: 
            cat = 'male_elder'
    return cat

titanic_df['combined_cat'] = titanic_df.apply(lambda row: get_combined_cat(row['sex'], row['age']), axis=1)
titanic_df.head()
```

```python
# 패밀리 변수 만들기
titanic_df['family'] = titanic_df['sibsp'] + titanic_df['parch']
titanic_df
```

```python
titanic_df.info()
```

# 파생변수 만들기완료.. 이후 인코딩작업

```python
# onehot인코딩 combined_cat
onehot_combined_cat = pd.get_dummies(titanic_df['combined_cat'], prefix='combined')
titanic_df = pd.concat([titanic_df, onehot_combined_cat], axis=1)
titanic_df.head()
```

```python
titanic_df.combined_cat.value_counts()
```

```python
titanic_df.fare_cat.value_counts()
```

```python
# 불필요한 컬럼 제거
def drop_feature(df):
    df.drop(['age', 'fare', 'sex', 'sibsp', 'parch', 'combined_cat'], axis=1, inplace=True)
    return df

def transform_features(df):
    df = drop_feature(df)
    return df

titanic_df = transform_features(titanic_df)
titanic_df.head()
```

```python
# 데이터셋을 훈련, 테스트로 구분하여 분리
from sklearn.model_selection import train_test_split

# 독립변수와 종속변수 분리
y_t_df = titanic_df['survived'] # 종속변수
X_t_df = titanic_df.drop('survived', axis=1) # 독립변수

# 학습용 데이터와 평가용 데이터를 8:2로 분리
X_train, X_test, y_train, y_test = train_test_split(X_t_df, y_t_df, test_size=0.2, random_state=11)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 랜덤 포레스트 모델 학습 및 평가
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, rf_pred).round(2)

# 로지스틱 회귀 모델 학습 및 평가
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test, lr_pred).round(2)

print('랜덤 포레스트 정확도: {}, 로지스틱 회귀 정확도: {}'.format(accuracy_rf, accuracy_lr))
```
