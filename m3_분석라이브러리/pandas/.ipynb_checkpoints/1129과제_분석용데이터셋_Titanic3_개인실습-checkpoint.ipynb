### 전처리 내역 예시
- pclass와 fare_cat중 분석변수 선택
- embarked 원핫인코딩

```python
# value_counts 로 컬럼 뽑기 함수 # unique 로 컬럼 뽑기 함수
def column_uniq_vcnt(column_name):
    print(f"컬럼 : {column_name}, 유니크값 : {titanic_df[column_name].unique()}")
    print(f"컬럼 : {column_name}, 밸류카운트값 : {titanic_df[column_name].value_counts()}")
```

```python
# 컬럼 age
column_uniq_vcnt('age')
```

```python
# 컬럼 cabin
column_uniq_vcnt('cabin')
```

```python
# 컬럼 fare
column_uniq_vcnt('fare')
```

```python
# 컬럼 embarked
column_uniq_vcnt('embarked')
```

```python
# 컬럼 home.dest
column_uniq_vcnt('body')
```

#### ○ null 처리

* Age : null값 263개를 갖고 있음
    * 전처리 : O
        * 위의 column 분석에 의하면 영향력이 적지 않으므로 null값을 적절히 처리한 후 활용할 것
* Cabin : null값 1014개를 갖고 있음
    * 전처리 : X(제외)
        * 사고 시각은 자정 즈음이었으나, 갑작스러운 사고로 배가 단숨에 침몰한 것이 아니므로 객실의 레벨과 생존여부는 관련이 없을 것으로 추측됨
        * 객실의 수준을 통해 얻을 수 있는 정보는 대표적으로 탑승객의 재력이며, 이는 Fare 등으로 대체할 수 있다고 여겨짐
* Fare : null값 1개 갖고 있음
    * 전처리 : O
        * null 값을 적절히 처리한 후 활용할 것. 티켓의 요금은 pclass 와 연관이 높을 것으로 봐서 적절히 값을 넣어주면 될듯
* Embarked : null값 2개를 갖고 있음 [거의 없음]
    * 전처리 : O
        * null값의 개수가 신경을 쓰지 않아도 될 정도로 적으므로 아예 삭제해도 무방할 것이며, 혹은 Name이나 동반자 여부를 통해 전처리의 신뢰성을 높일 수 있을 것이라고 생각됨.
* boat, body, home.dest : 널 값이 상당히 많음
    * 전처리 : X(제외)
        * 생존 여부 모델에는 큰 관련이 없을 것으로 추측됨. 영향력이 적은 컬럼으로 제외.

```python
# fare 가 널값의 개수가 1개여서 이거먼저해보자.
# fare 에서 널값이 있는 행
titanic_df[titanic_df.fare.isna()]
```

```python
# fare Null에 대체값 탐색
# fare 는 pclass 와 관련이 있을 것으로 보고 pclass가 3인 것을 좀더 탐색
pclass_3_df = titanic_df[titanic_df['pclass'] == 3]
pclass_3_df
# pclass가 3인 사람들의 fare 의 describe 로 확인하면 될듯
pclass_3_df.fare.describe()
# 최댓값이 평균값과 사분할 값보다 너무 격차가 커서
# 임의로 범위를 정한뒤 탐색후에 값을 정해야 할것으로 보임
# 총708, min0, max69, 평균13, std11, 25%7, 50%8, 75%15
# 0부터, 75% 값에 mean을 더한값까지를 범위로 하고 다시 탐색.
pclass_3_df[pclass_3_df['fare'] <= 13.302889+15.245800].fare.describe()
# mean 값은 10.420884, 분포가 25퍼가 7.몇, 50퍼가 7.몇 으로 가장 많아서
# 빈도수가 많은 50% 의 값인 7.925000(7.925) 으로 널값을 대체하기로 함
```

```python
# fare Null 값처리 : 10.420884 으로 대체
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
titanic_df.fare.fillna(7.925, inplace=True)
titanic_df[titanic_df.fare.isna()]
```
