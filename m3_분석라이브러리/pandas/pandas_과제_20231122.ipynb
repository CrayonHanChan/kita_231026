### [과제 2] df_4에 대하여 다음을 수행하세요.
- 첫번째 컬럼을 출력하세요(2가지 방법)
- 두번째 행을 출력하세요.

```python
# 과제 2번을 위한 Dataframe 생성

import pandas as pd
import numpy as np

# DataFrame 2개 생성

df_1 = pd.DataFrame({
    'col1':np.array([1,2,3]),
    'col2':np.array(['A','B','C'])
})
df_2 = pd.DataFrame({
    'col1':np.array([4,5,6]),
    'col2':np.array(['D','E','F'])
})

# '행'방향으로 병합하여 df_4 생성
df_4 = pd.concat([df_1, df_2], axis=0)   # axis=0 :'행' 방향 (기본값이므로 생략 가능.)
```

```python
df_4
```

```python
# 첫번째 컬럼 출력 
print(df_4.loc[:,'col1'], "\n")        # .loc() 사용 : index 이름 사용
print(df_4.iloc[:,0], "\n")            # .iloc() 사용 : index 번호 사용


# 두번째 행 출력
print(df_4.iloc[1,:])
```
